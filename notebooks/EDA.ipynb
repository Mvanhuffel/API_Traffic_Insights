```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set style of plots
sns.set_style("whitegrid")

df = pd.read_csv('../data/raw/MOCK_DATA.csv')
```

```python
# BASIC INFORMATION

# First few rows
display(df.head())
# Summary of the dataframe
display(df.info())
# Number of missing values in each column
display(df.isnull().sum())
```

```python
# SUMMARY STATISTICS

# Summary stats for numerical col
display(df.describe())

# Distribution of categories
categorical_columns = ['customer', 'instance_name', 'uri', 'servicename', 'remote_name', 'status', 'final_status', 'transaction_date', 'partner']

for col in categorical_columns:
    display(f"Value counts for {col}:")
    display(df[col].value_counts())
    print("\n")
```

```python
# DATA TYPES

display(df.dtypes)
```

```python
# HISTOGRAMS

# Plot histograms for all numerical columns
df.hist(bins=20, figsize=(14, 10))
plt.tight_layout()
plt.show()
```

```python
# BOXPLOTS FOR NUMERICAL DATA

# Select only numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Setup for subplot dimensions
plots_per_row = 3 
num_rows = len(numerical_cols) // plots_per_row + (len(numerical_cols) % plots_per_row > 0)

plt.figure(figsize=(18, num_rows * 4))

for i, col in enumerate(numerical_cols):
    plt.subplot(num_rows, plots_per_row, i + 1)  # Create a subplot for each numerical column
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')

plt.tight_layout()
plt.show()
```

```python
# COUNT PLOTS FOR CATEGORICAL DATA

import matplotlib.pyplot as plt
import seaborn as sns

# Select only categorical columns, excluding 'bool'
categorical_cols = df.select_dtypes(include=['object']).columns

# Define top N categories
top_n = 10

# Setup for subplot dimensions
plots_per_row = 3
num_rows = len(categorical_cols) // plots_per_row + (len(categorical_cols) % plots_per_row > 0)

plt.figure(figsize=(18, num_rows * 5))  # Adjust figure size as needed

for i, col in enumerate(categorical_cols):
    plt.subplot(num_rows, plots_per_row, i + 1)
    
    # Get the counts of each category and select the top N
    top_categories = df[col].value_counts().head(top_n).index
    
    # Filter the DataFrame for the top N categories
    df_filtered = df[df[col].isin(top_categories)]
    
    # Create the count plot
    sns.countplot(data=df_filtered, x=col, order=top_categories)
    plt.title(f'Top 10 {col} by Count')
    plt.xticks(rotation=45)  # Rotate labels for better readability
    plt.tight_layout()

plt.show()
```

```python
# CORRELATION HEATMAP FOR NUMERICAL DATA

# Calculate the correlation matrix for numerical columns
corr_matrix = df[numerical_cols].corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar_kws={'shrink': .8}, square=True, linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```
